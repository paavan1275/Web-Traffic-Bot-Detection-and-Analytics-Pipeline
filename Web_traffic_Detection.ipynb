```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
seman = pd.read_csv("semantic_features.csv")
simple = pd.read_csv("simple_features.csv")
```

```python
seman1 = pd.DataFrame(seman)
simple1 = pd.DataFrame(simple)
```

```python
seman1.shape
```

```python
simple1.shape
```

```python
data = pd.merge(seman1, simple1, on = "ID", how = "inner", suffixes = ("_1", "_2"))
```

```python
#Data Cleaning
```

```python
data = data.drop(columns = ["ROBOT_1"])
```

```python
data.isnull().sum()
```

```python
#Filling NaN values with value "0"
```

```python
data1 = data.fillna(0)
```

```python
data1
```

```python
data1.isna().sum()
```

```python
data1.shape
```

```python
#Splitting the dataframe for x and y
```

```python
x_data = data1.iloc[:,1:-1]
```

```python
y_data = data1.iloc[:,-1]
```

```python
print("shape of x_data : ",x_data.shape)
print("shape of y_data : ",y_data.shape)
```

```python
#Splitting the dataset for the Model training [train - test : (80 - 20)]
```

```python
#Importing the requried module to split the data
from sklearn.model_selection import train_test_split

#Splitting the data
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.2, random_state = 42)
```

```python
print("x training data : ", x_train.shape)
print("y training data : ", y_train.shape)
print("-"*40)
print("x testing data : ", x_test.shape)
print("y testing data : ", y_test.shape)
```

```python
x_train.head()
```

```python
#Performing the maximum models to get the best accuracy
# 1. XGBoost
# 2. LightGBM
```

```python
# 1. XGBoost model
# Accuracy
# XGBoost Acuracy: 0.977699354257396
#               precision    recall  f1-score   support

#            0       0.99      0.99      0.99     10590
#            1       0.95      0.94      0.95      2728

#     accuracy                           0.98     13318
#    macro avg       0.97      0.97      0.97     13318
# weighted avg       0.98      0.98      0.98     13318
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


xgb_model = XGBClassifier(n_estimators=300, max_depth=6, learning_rate=0.05, subsample=0.8, colsample_bytree=0.8, objective="binary:logistic", eval_metric="auc", n_jobs=-1, tree_method="hist")

xgb_model.fit(x_train, y_train)

y_predict = xgb_model.predict(x_test)
xgb_auc = accuracy_score(y_test, y_predict)
print("XGBoost Acuracy:", xgb_auc)

print(classification_report(y_test, y_predict))
```

```python
#LightGBM
#Acuracy of the model
# LightGBM AUC: 0.9783000450518096
#               precision    recall  f1-score   support

#            0       0.99      0.99      0.99     10590
#            1       0.94      0.95      0.95      2728

#     accuracy                           0.98     13318
#    macro avg       0.97      0.97      0.97     13318
# weighted avg       0.98      0.98      0.98     13318
```

```python
from lightgbm import LGBMClassifier

lgb_model = LGBMClassifier(n_estimators=300, learning_rate=0.05, max_depth=-1, num_leaves=31, subsample=0.8, colsample_bytree=0.8, objective="binary", n_jobs=-1)

lgb_model.fit(x_train, y_train)

y_pred_lgb = lgb_model.predict(x_test)
lgb_auc = accuracy_score(y_test, y_pred_lgb)
print("LightGBM AUC:", lgb_auc)

print(classification_report(y_test, y_pred_lgb))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
ran_model = RandomForestClassifier()

ran_model.fit(x_train, y_train)
```

```python
y_pred = ran_model.predict(x_test)

ran_accuracy = accuracy_score(y_test, y_pred)

print("Random Forest accuracy: ", ran_accuracy)
print(classification_report(y_test, y_pred))
```
